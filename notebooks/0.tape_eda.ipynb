### Fluorescence

Mutations centered around WT Green Fluorescence Protein (GFP). The sequence corresponding to WT GFP is the sequence which has zero mutations in the training set. Deletions are counted as mutations. The num_mutations field in the original TAPE dataset tracks edit distance from WT (substitutions + deletions), not substitutions only. A 236 AA sequence with num_mutations=1 has one deletion and nothing else; a 235 AA sequence with num_mutations=2 has exactly two deletions.

```python
train_df = pd.read_csv(".data/fluorescence/train.csv")
```

```python
train_df["sequence"].map(lambda x: len(x)).value_counts()
```

```python
train_df["target"].hist(bins=100)
```
